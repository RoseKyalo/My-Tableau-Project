```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# Initialize an empty list to store data from all pages
all_data = []

# Iterate over each page
for page_number in range(1, 19):
    url = f'https://www.missingvoices.or.ke/statistics?search=&order=field_location_city&sort=asc&page={page_number}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    flat_data = []
    data_boxes = soup.find('table', class_='views-table')
    for data_box in data_boxes:
        soup1 = BeautifulSoup(data_box.text, 'html.parser')
        soup1.find(class_='colorbox-load')
        text_data = soup1.text
        # Split the text into lines and then split each line by whitespace to create a list of lists
        rows = [line.split() for line in text_data.strip().split('\n')]
        flat_data.append(rows)

    # Concatenate all the inner lists into a single list
    flat_data = [item for sublist in flat_data for item in sublist]
    flat_data = flat_data[15:]
    processed_data = [flat_data[i:i+11] for i in range(0, len(flat_data), 11)]

    # Append the processed data for the current page to the list of all data
    all_data.extend(processed_data)

# Create a DataFrame from the combined data
df = pd.DataFrame(all_data, columns=['No', 'Name', 'Age', 'Sex', 'Location', 'Manner_of_Death',
                                      'Date of Incident', '[]', 'Sort descending', 'Profile Image',
                                      '[]'])

# Drop the specified columns from the DataFrame
columns_to_drop = ['No', '[]', 'Sort descending', 'Profile Image', '[]']
df.drop(columns=columns_to_drop, inplace=True, axis=1)

# Apply the regular expression to clean up column values
for column in df.columns:
    df[column] = df[column].apply(lambda x: ' '.join(re.findall(r'[a-zA-Z0-9]+', str(x))))
```
